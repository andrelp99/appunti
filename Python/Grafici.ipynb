# Boxplot

```python
# Boxplot semplice
# linewidth per la larghezza delle righe, notch per presenza/assenza del notch, width per larghezza boxplot (tutti facoltativi)

sns.boxplot(x='giorno', y='scontrini', data=rist1, linewidth=2, notch=True, width=0.3)
plt.title("Boxplot", loc="left")
plt.show()
```

```python
# Boxplot con jitter points
# colore definisce i colori dei pallini, jitter quanto sono distribuiti in larghezza nei vari boxplot
# size la dimensione

sns.boxplot(x='giorno', y='scontrini', data=rist1)
sns.stripplot(x='giorno', y='scontrini', data=rist1, color="orange",  jitter=1.2, size=2.5)
plt.title("Boxplot", loc="left")
plt.show()
```

```python
# Boxplot semplice con statistiche su grafico

ax = sns.boxplot(x='giorno', y='scontrini', data=rist1)

medians = rist1.groupby(['giorno'])['scontrini'].median().values
nobs = rist1.groupby("giorno").size().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n: " + i for i in nobs]

pos = range(len(nobs))
for tick,label in zip(pos,ax.get_xticklabels()):
    plt.text(pos[tick], medians[tick] + 0.4, nobs[tick], horizontalalignment='center', size='medium', color='b', weight='semibold')

plt.title("Boxplot", loc="left")
plt.show()

```

```python
# Boxplot semplice con stile grafico predefinito
# color facoltativo, per definire un colore prevalente
# palette per una palette prevalente

sns.set(style="darkgrid")
sns.boxplot(x='giorno', y='scontrini', data=rist1, color='skyblue', palette="Blues")
plt.title("Boxplot", loc="left")
plt.show()
```

```python
# Boxplot semplice con stile grafico predefinito
# palette per una palette definita dall'utente

my_pal = {"Saturday": "gray", "Sunday": "blue", "Monday":"yellow", "Tuesday": "red", "Wednesday": "green", "Thursday":"purple",  "Friday":"orange"}
sns.boxplot(x='giorno', y='scontrini', data=rist1, palette=my_pal)
plt.title("Boxplot", loc="left")
plt.show()
```

```python
# Boxplot semplice con stile grafico predefinito
# palette con una classe in risalto definita dall'utente

my_pal = {giorno: "red" if giorno == "Monday" else "blue" for giorno in rist1.giorno.unique()}
sns.boxplot(x='giorno', y='scontrini', data=rist1, palette=my_pal)
plt.title("Boxplot", loc="left")
plt.show()
```

```python
# DA PROVARE

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

def plot_correlation_from_dataframe(dataframe, columns=None, annot=True):
    '''
        Plots the correlations between the columns of a dataframe

        Attributes:
            - dataframe     (pandas.DataFrame) : the dataframe
            - columns       ([string])         : the columns to analyze, if None or []: all the columns 
            - values        (bool)             : whether or not show the correlation values in the plot
    '''

    if columns is None or len(columns) == 0:
        df = dataframe
    else:
        df = dataframe[columns].copy()

    colormap = plt.cm.RdBu_r
    plt.figure(figsize=(30,20), facecolor='white')
    plt.title(u'Correlation ', y=1.05, size=16)

    mask = np.zeros_like(df.corr())
    mask[np.triu_indices_from(mask)] = True

    svm = sns.heatmap(df.corr(), mask=mask, linewidths=0.1,vmax=1.0,square=True, center=0, cmap=colormap, linecolor='white', annot=annot)
```
